```python
a=float(input("Enter the cost of first item:"))
b=float(input("Enter the cost of second item:"))
c=float(input("Enter the cost of third item:"))
cost=a+b+c
print("Total cost=",cost)
if cost>50:
  cost=cost-cost*0.1
print("Cost after discount=",cost);
```
